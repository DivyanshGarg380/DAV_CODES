#### 2) For each of the numeric attributes (proteins upto vitamins) ,identify and replace all missing data(indicated with -1) with the arithmetic mean of the attribute

```python
#removing missing values(-1 values)
cereals_df = cereals_df.replace(-1, np.nan).fillna(cereals_df.mean(numeric_only=True))
```

#### 3) . Create a table with the 5 number summary of all the numeric attributes after treating missing values.

#### 4) For each numeric attribute (proteins upto vitamins), identify and replace all noisy data with the median of attribute.

```python
cereals_df.plot(kind='box',title = "Box Plot")
```

```python

target_columns = cereals_df.loc[:, 'protein':'vitamins'].columns

# Detect and replace noise with median
for col in target_columns:
    Q1 = cereals_df[col].quantile(0.25)
    Q3 = cereals_df[col].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate bounds for noise
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Calculate median value
    median_value = cereals_df[col].median()

    # Replace noisy values with median
    cereals_df[col] = cereals_df[col].apply(lambda x: median_value if x < lower_bound or x > upper_bound else x)
```

##### 13)  How does the sugar content compare in cereals with high vs. low ratings?

```python
median_rating = cereals_df['rating'].median()
cereals_df['rating_category'] = cereals_df['rating'].apply(lambda x: 'High' if x >= median_rating else 'Low')
sugar_comparison = cereals_df.groupby('rating_category')['sugars'].mean()
sugar_comparison
```

##### 14) Do cereals marketed as "healthy" (e.g., high fiber, low sugar) tend to have higher ratings?

```python
fiber_median = cereals_df['fiber'].median()
sugar_median = cereals_df['sugars'].median()


cereals_df['health_category'] = cereals_df.apply(
    lambda x: 'Healthy' if x['fiber'] >= fiber_median and x['sugars'] <= sugar_median else 'Unhealthy',
    axis=1
)

rating_comparison = cereals_df.groupby('health_category')['rating'].mean()

rating_comparison
```
