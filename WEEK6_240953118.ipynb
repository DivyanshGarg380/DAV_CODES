##### Q1 For the MTCARS dataset, answer the specified questions with summarization and effective visuals

```python
car_df['mpg'].plot(kind='hist', bins=10, alpha=0.7, edgecolor='black', figsize=(8, 6), legend=True)
plt.title('Distribution of Fuel Efficiency (MPG) of Cars')
plt.xlabel('Miles Per Gallon (MPG)')
plt.ylabel('Frequency')
```

##### Q8 What is average MPG for manual vs. automatic cars?

```python
dat_man = car_df[car_df["am"] == 1]["mpg"]
print("Avg of manual car : ",dat_man.mean())
dat_auto = car_df[car_df["am"] == 0]["mpg"]
print("Avg of auto car : ",dat_auto.mean())
```

##### Q9  Draw Side by Side box plot to understand the difference in fuel efficience of Manual vs Automatic cars.  Analyze and write about fuel efficiency in each group (manual vs. automatic).

```python
car_df.boxplot(column='mpg', by='am', grid=False, figsize=(8, 6))


plt.title('Fuel Efficiency (MPG) by Transmission Type')
plt.xlabel('Transmission (0 = Automatic, 1 = Manual)')
plt.ylabel('Miles Per Gallon (MPG)')
```

##### Q10 What is the relationship between the weight of the car and MPG? Create scatter plot

```python
car_df.plot(kind='scatter', x='wt', y='mpg', color='blue', figsize=(8, 6), title='Relationship Between Car Weight and MPG')

plt.xlabel('Weight of Car (1000 lbs)')
plt.ylabel('Miles Per Gallon (MPG)')
```

```python
dat_x = car_df[["mpg","hp"]]
dat_x.plot(kind='scatter',x='hp',y='mpg',color='blue',title='Scatter plot from dataframe')
```

##### Q11 Categorize the cars based on the number of gears in the cars. How many cars are there in each type?

```python
gear_counts = car_df['gear'].value_counts()
gear_counts
```

##### Q12  What is the relationship between fuel efficiency and the number of gears in the car?

```python
car_df.boxplot(column='mpg', by='gear', grid=False, figsize=(8, 6))

plt.title('Fuel Efficiency (MPG) by Number of Gears')
plt.xlabel('Number of Gears')
plt.ylabel('Miles Per Gallon (MPG)')
```

##### Q13 Explain the relationship between horse power and number of cylinders in the car.

```python
car_df.plot(kind='scatter', x='hp', y='cyl', color='purple', figsize=(8, 6), title='Relationship Between Horsepower and Cylinders')

plt.xlabel('Horsepower (hp)')
plt.ylabel('Number of Cylinders')
```

##### Q14 Explain the relationship between displacement and gross horse power.

```python
car_df.plot(kind='scatter', x='disp', y='hp', color='orange', figsize=(8, 6), title='Relationship Between Displacement and Gross Horsepower')

plt.xlabel('Displacement (cu.in.)')
plt.ylabel('Gross Horsepower (hp)')
```
