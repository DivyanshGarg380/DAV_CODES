```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("deliveries.csv")
df
```

```python
# 1. Find and visualize the total runs scored by each team.
runs = df.groupby('batting_team')['total_runs'].sum().reset_index()
runs = runs.sort_values(by='total_runs', ascending=False)
runs
```

```python
# plot for Q1 ( doing a bar graph :) )
plt.figure(figsize=(6,4))
sns.barplot(x='total_runs', y='batting_team', data=runs, palette='viridis')
plt.xlabel("Total Runs Scored")
plt.ylabel("Team")
plt.title("Total Runs Scored by Each Team")
plt.show()
```

Observation:   
1) Most runs are made by Mumbai Indians
2) Kochi Tusker Kerala made the least over the season

```python
# 2. Identify and display the names of the top 10 highest-scoring batters with their total runs scored
topb = df.groupby('batter')['batsman_runs'].sum().reset_index()
topb = topb.sort_values(by='batsman_runs', ascending=False).head(10)
topb
```

Observation :   
1) Top batter is Virat Kohli
2) KD karthik is at the bottom this season

```python
# 3. Find and visualize the top 10 bowlers with the most wickets
top_bowlers = df[df['dismissal_kind'].notna()]
top_bowlers = top_bowlers.groupby('bowler')['dismissal_kind'].count().reset_index()
top_bowlers = top_bowlers.sort_values(by='dismissal_kind', ascending=False).head(10)
plt.figure(figsize=(6,4))
sns.barplot(x='dismissal_kind', y='bowler', data=top_bowlers, palette='magma')
plt.xlabel("Total Wickets Taken")
plt.ylabel("Bowler")
plt.title("Top 10 Bowlers with Most Wickets")
plt.show()
```

Observation :   
1) Top bowler who took the max wickets this season is YS Chahal
2) RA Jadeja took around only 175 wickets

```python
# 4. Tabulate and visualize the total number of extras conceded by each team.
extras = df.groupby('bowling_team')['extra_runs'].sum().reset_index()
extras
```
