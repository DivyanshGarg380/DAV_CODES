```python
# Step 4: Text(s) with maximum likes
max_likes = df["Likes"].max()
most_liked_texts = df[df["Likes"] == max_likes][["Text", "Likes"]]
most_liked_texts
```

```python
# Step 5: Text(s) with maximum reposts and their sentiment
max_reposts = df["Retweets"].max()
most_reposted_texts = df[df["Retweets"] == max_reposts][["Text", "Retweets", "Sentiment"]]
most_reposted_texts
```

```python
# Step 6: Top 5 countries posting the most positive texts
positive_texts = df[df["Sentiment"] == "positive"]
top_countries = positive_texts["Country"].value_counts().nlargest(5)

plt.figure(figsize=(8, 5))
sns.barplot(x=top_countries.index, y=top_countries.values, palette="Blues_r")
plt.title("Top 5 Countries Posting Most Positive Texts")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

Observation:  
1) USA has the most positive texts

```python
# Step 7: Texts with 'Travel' in hashtags
travel_texts = df[df["Hashtags"].str.contains("travel", case=False, na=False)][["Text", "Hashtags"]]
travel_texts
```

Observation:  
1) A total of 7 texts contain "travel" hashtag in them

```python
# Step 8: Top 5 most liked negative sentiment posts from 2020
negative_2020 = df[(df["Year"] == 2020) & (df["Sentiment"] == "negative")]
top_negative_2020 = negative_2020.nlargest(5, "Likes")[["Text", "Likes"]]
top_negative_2020
```

```python
# Step 9: Users and posts with 'Happiness' sentiment
happiness_posts = df[df["Sentiment"] == "happiness"][["User", "Text"]]
happiness_posts
```

```python
# Step 10: Years and countrywide tweets for sentiments Positive, Negative, and Neutral
filtered_df = df[df["Sentiment"].isin(["positive", "negative", "neutral"])]
year_country_sentiments = filtered_df.groupby(["Year", "Country", "Sentiment"]).size().unstack(fill_value=0)

plt.figure(figsize=(12, 6))
year_country_sentiments.plot(kind="bar", stacked=True, figsize=(12, 6), colormap="viridis")
plt.title("Sentiments Over Years and Countries")
plt.xlabel("Year, Country")
plt.ylabel("Tweet Count")
plt.legend(title="Sentiment")
plt.show()
```

Observation:  
1) Most positive sentiment was seen in 2023 - India
2) Most neutral sentiment was seen in 2023 - Canada
3) Most negative sentiment was seen in 2023 -  Canada
