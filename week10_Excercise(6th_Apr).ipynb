```python
product_data = pd.merge(products, categories, on="CategoryID", how="left")
order_data = pd.merge(orders, customers, on="CustomerID", how="left")
```

```python
print("Missing values in Products:\n", products.isnull().sum(), "\n")
print("Missing values in Categories:\n", categories.isnull().sum(), "\n")
print("Missing values in Customers:\n", customers.isnull().sum(), "\n")
print("Missing values in Orders:\n", orders.isnull().sum(), "\n")
```

```python
customers["Region"].fillna("Unknown", inplace=True)
orders["ShipRegion"].fillna("Unknown", inplace=True)
customers["PostalCode"].fillna("00000", inplace=True)
customers["Fax"].fillna("Not Provided", inplace=True)
orders["ShipPostalCode"].fillna("00000", inplace=True)
orders.drop(columns=["ShippedDate"], inplace=True)
```

```python
dup_products = products.duplicated().sum()
dup_categories = categories.duplicated().sum()
dup_customers = customers.duplicated().sum()
dup_orders = orders.duplicated().sum()

print("🔁 Duplicate Rows:")
print(f"Products: {dup_products}")
print(f"Categories: {dup_categories}")
print(f"Customers: {dup_customers}")
print(f"Orders: {dup_orders}")
```

```python
# Lowercase for orders_data
for col in order_data.select_dtypes(include='object').columns:
    order_data[col] = order_data[col].str.lower()

# Lowercase for product_data
for col in product_data.select_dtypes(include='object').columns:
    product_data[col] = product_data[col].str.lower()
```
