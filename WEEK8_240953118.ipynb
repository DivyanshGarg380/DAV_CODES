```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("mtcars.csv") 
pd.set_option('display.max_columns',100) # display all coloumns
pd.set_option('display.max_rows',None) # display all rows
```

```python
# View the data set
df
```

```python
# Create cross tabulation using pandas. Cross tabulation features will help find occurences
# the combination of values of two coloumns
# Tabulate the cylinders and gear cars
tab1 = pd.crosstab(df['cyl'],df['gear'])
print("Table summarizing number of cars (Cylinder vs Gears)")
print(tab1)
```

Observation :
1) 6 cylinder cars are less   
2) 4 gears cylinder cars are not present at the moment    
3) Most commmon type of car is gear 3 cylinder - 8   
4) Most popular car is cylinder - 8 

```python
# Heatmap of the cross tab result . The best way to visualize a cross tab is to use a heatmap
sns.heatmap(tab1,annot=True,fmt='d',cmap='YlGnBu')
plt.title("Cross tab Heatmap: Cylinders vs Gears")
plt.xlabel("Gears")
plt.ylabel("Cylinders")
plt.gca().collections[0].colorbar.set_label("Frequency")
```

Observation :  
1) Cars with 8 cylinders and 3 gears are 12 in total and is the highest in frequency.   
2) Cars with the highest frequency are marked in purple colour and the one's with lowest frequecny are marked with the     colour light yellow.    
3) No such car is present with 6 cylinders and 4 gears 

```python
# Sort values() takes the coloumn names , based on which the records need to be sorted.
# By default , sorting is done in ascending order
tab2 = df[['model','mpg']].sort_values('mpg',ascending=False) #True if wanna sort in ascending
print("Table displaying Car models (Best to worst Fuel efficiency)")
print(tab2)
```

```python
# Group by and aggregate: Average MPG by Numbers of Cylinders
tab3 = df.groupby('cyl')['mpg'].mean()
print("Table displaying Number of Cylinders & Average mpg")
print(tab3)
```

Observation :   
1) mpg is the most when using 4 cylinders   
2) mpg is the least when using 8 cylinders   

```python
# Bar chart is a frequency chart for qualitative variable (or categorial variable).
# Bar chart can be used to assses the most-occuring and Least-occuring categories within a dataframe
sns.barplot(x='cyl',y='mpg',data=df,ci=None)
```

Observation:  
1) More mpg for less cylinders used.  
2) 25 mpg would be given for just 4 cylinders if used.   
3) Only 15 mpg would be given if 8 cylinders are used.   
4) 6 cylinders offer decent mpg , around 20

```python
# Scatter plot of mpg vs hp
plt.figure(figsize=(6,4))
sns.scatterplot(x='hp',y='mpg',data=df)
plt.title("MPG vs HP")
plt.show()
```

Observation:  
1) It can be infered that more mpg is given if a car has less hp.   
2) With hp > 300 , mpg is approximately 15 --> Average.   
3) With hp > 50 and hp < 150 , mpg is between 15 to 35 ---> Good output.   
4) A car with 200 hp would have the least mpg , approximately 10 ---> Not good. 

```python
# Histogram plot of mpg
plt.figure(figsize=(6,4))
sns.histplot(df['mpg'],bins=10,kde=True)
plt.title("Distribution of MPG")
plt.show()
```

Observation:  
1) The lines represents the density.  
2) Which means Higeher density is towards to left and least density is towards to right.  
3) Right skewed

```python
# Pair plot of numerical columns 
sns.pairplot(df[['mpg','hp','wt','qsec']])
plt.show()
```

```python
# Violin plot of mpg by cylinder count
plt.figure(figsize=(6,4))
sns.violinplot(x='cyl',y='mpg',data=df)
plt.title("MPG distribution by Cylinder count")
plt.show()
```

Observation:   
1) The white dot in middle of the graphs is called the median.    
2) Data is concentrated.  

```python
# Regression plot of mpg vs weight
plt.figure(figsize=(6,4))
sns.regplot(x='wt',y='mpg',data=df)
plt.title("MPG vs Weight regression plot")
plt.show()
```

Observation:  
1) Negative co-relation since graph is downwards. 

```python
# Swarm plot of mpg by cylinder count
plt.figure(figsize=(6,4))
sns.swarmplot(x='cyl',y='mpg',data=df)
plt.title("MPG vs Cylinder count by Swarm plot")
plt.show()
```

Observation:   
1) As the dots are concentrated for cars having 6 cylinders. Not much variation observed.   
2) Minor variations are seen in cars having 8 cylinders.   
3) Large variations are seen in cars having 4 cylinders ---> Dots are scattered.
