```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("matches.csv")
df
```

```python
# Find the names of teams that are part of IPL.
teams = df["team1"].unique()
print("Teams that are part of IPL:")
print(teams)
```

```python
# We extracted unique team names from the team1 and team2 columns.
# This gives a complete list of teams that have participated in the IPL over the years.
# This is useful for understanding the competition's history and tracking team participation.
```

```python
# Visualize the total number of matches won by each team.
team_wins = df["winner"].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=team_wins.values, y=team_wins.index, palette="viridis")
plt.xlabel("Total Matches Won")
plt.ylabel("Teams")
plt.title("Total Matches Won by Each Team")
plt.show()
print("Total Matches Won by Each Team:")
print(team_wins)
```

```python
# The bar chart shows the number of matches won by each team.
# Teams with the highest bars have historically been the most successful.
# If a team has significantly fewer wins, it might indicate poor performance across seasons.
# The visualization helps identify the dominant teams in IPL history.
```

```python
# 3. List the names of umpires who have served as third umpires and the number of matches.
second_umpires = df["umpire2"].value_counts()
print("Second Umpires and Number of Matches:")
print(second_umpires)
```

```python
# We counted how many times each umpire has served as the second umpire.
# The umpires with the highest counts are the most experienced in reviewing decisions.
# Understanding umpire assignments can help analyze their decision accuracy and impact.
```

```python
# 4. Visualize how many matches were held in each venue.
venue_counts = df["venue"].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=venue_counts.values, y=venue_counts.index, palette="coolwarm")
plt.xlabel("Total Matches Held")
plt.ylabel("Venue")
plt.title("Total Matches Held at Each Venue")
plt.show()
print("Total Matches Held at Each Venue:")
print(venue_counts)
```

```python
# The graph indicates which venues have hosted the most IPL matches.
# Some venues, like Wankhede Stadium or Eden Gardens, might have more matches due to their popularity.
# If a venue has hosted only a few matches, it could be a newly added or less frequently used stadium.
# This insight helps in understanding venue selection and tournament expansion.
```

```python
# 5. Visualize the top 10 players identified as the player of the match for the most matches.
top_players = df["player_of_match"].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_players.values, y=top_players.index, palette="magma")
plt.xlabel("Total Matches Won as Player of the Match")
plt.ylabel("Player")
plt.title("Top 10 Players with Most Player of the Match Awards")
plt.show()
print("Top 10 Players with Most Player of the Match Awards:")
print(top_players)
```

```python
# The graph highlights players who have consistently performed well.
# Players at the top are likely match-winners who have significantly impacted their teams.
# If a player has received many awards, it suggests they played crucial roles in victories.
# This helps recognize key performers in IPL history.
```

```python
# 6. Visualize the bottom 10 venues where the least number of matches were played.
bottom_venues = venue_counts.tail(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=bottom_venues.values, y=bottom_venues.index, palette="OrRd")
plt.xlabel("Total Matches Held")
plt.ylabel("Venue")
plt.title("Bottom 10 Venues with Least Matches Played")
plt.show()
print("Bottom 10 Venues with Least Matches Played:")
print(bottom_venues)
```

```python
# Some venues have hosted very few matches, indicating limited use.
# Possible reasons:
# These could be neutral or international venues.
# They might have been used only for specific seasons.
# Some could be removed from IPL hosting lists due to infrastructure or other factors.
# This visualization helps in understanding IPL’s venue strategy.
```

```python
# 7. Visualize the number of wins of each team at 'MA Chidambaram Stadium, Chepauk'.
chidambaram_wins = df[df["venue"] == "MA Chidambaram Stadium, Chepauk"]["winner"].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=chidambaram_wins.values, y=chidambaram_wins.index, palette="BuGn")
plt.xlabel("Total Wins")
plt.ylabel("Teams")
plt.title("Wins at MA Chidambaram Stadium, Chepauk")
plt.show()
print("Wins at MA Chidambaram Stadium, Chepauk:")
print(chidambaram_wins)
```

```python
# The chart shows which teams have performed best at Chepauk.
# Some teams may have a home advantage and a significantly higher number of wins.
# If a team has struggled at this venue, it could be due to pitch conditions (e.g., spin-friendly pitches).
# This helps in analyzing venue-based team strengths.
```

```python
# 8. Find and list the venues where the match resulted in a tie.
tie_venues = df[df["result"] == "tie"]["venue"].unique()
print("Venues where matches resulted in a tie:")
print(tie_venues)
```

```python
# A tie means both teams scored the same runs, leading to a Super Over.
# If certain venues have more tied matches, it might indicate:
# Balanced pitch conditions.
# High competition at that stadium.
# Unpredictable gameplay at those venues.
# Understanding tie occurrences can provide insights into match excitement levels.
```

```python
# 9. List and visualize the best teams (top 5), season-wise.
season_wins = df.groupby(["season", "winner"]).size().reset_index(name="wins")
top_teams_per_season = season_wins.sort_values(by=["season", "wins"], ascending=[True, False]).groupby("season").head(5)
plt.figure(figsize=(12, 6))
sns.barplot(data=top_teams_per_season, x="season", y="wins", hue="winner", palette="Set2")
plt.xlabel("Season")
plt.ylabel("Total Wins")
plt.title("Top 5 Best Teams Season-wise")
plt.legend(title="Teams", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
print("Top 5 Best Teams Season-wise:")
print(top_teams_per_season)
```

```python
# Each season, the top 5 teams with the most wins are identified.
# This visualization shows how different teams have dominated different seasons.
# Some teams may consistently appear at the top, indicating sustained performance.
# Others may have had a one-season peak and then declined.
# This helps in tracking team performances over IPL seasons.
```

```python
b# 10. Analyze and visualize the relationship between winning the toss and winning the match.
df["toss_win_match_win"] = df["toss_winner"] == df["winner"]
toss_win_match_win_counts = df["toss_win_match_win"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(toss_win_match_win_counts, labels=["Won Match", "Lost Match"], autopct="%1.1f%%", colors=["lightgreen", "red"], startangle=90)
plt.title("Impact of Winning the Toss on Match Outcome")
plt.show()
print("Impact of Winning the Toss on Match Outcome:")
print(toss_win_match_win_counts)
```

```python
# If the percentage of "Won Toss & Won Match" is high, winning the toss gives a strong advantage.
# If the percentage is nearly 50-50, the toss has little influence on match outcomes.
# Some seasons or venues might show a stronger toss advantage due to:
# Dew factor affecting bowling in second innings.
# Pitch conditions favoring a particular strategy.
# This analysis is crucial for understanding match-winning strategies.
```
